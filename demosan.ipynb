```python

import pandas as pd
import os 
```

```python
files = [file for file in os.listdir('./videos_data')]
all_videos_data = pd.DataFrame()

for file in files:
   df = pd.read_csv("./videos_data/" + file)
   all_videos_data = pd.concat([all_videos_data, df])

all_videos_data.to_csv("all_videos_data.csv",index = False)
```

```python
all_videos_data = pd.read_csv("all_videos_data.csv")
all_videos_data.info()
```

```python
files = [file for file in os.listdir('./category_data')]
all_category_data = pd.DataFrame()

for file in files:
   df = pd.read_csv("./category_data/" + file)
   all_category_data = pd.concat([all_category_data, df])

all_category_data.to_csv("all_category_data.csv",index = False)
```
